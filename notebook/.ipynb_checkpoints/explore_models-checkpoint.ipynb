```python
def plot(ax, x, model, **kwargs):
    pars = model_fitters.get_pars(model, with_alias=True)
    y=pars['func'](x, *pars['bestv'])
    ax.plot(x,y,**kwargs)
```

### More details about Arnett model

```python
from sdapy import models, constants, functions
```

##### Tail model, t0 (slope) vs Mej (Ekin, with a fixed v)

```python
# velocity
vej = 8 # 1000 km/s

fig,ax1 = plt.subplots(1,1)
ax1.set_title('velocity=%.2f (1e3 km/s)'%vej)
ax2 = ax1.twinx()
ax3 = ax1.twiny()
for t0 in np.arange(40,110,1): # different tail slopes
    mej, ek = models.arnett_tail.t0_to_Mej_Ek(t0, vej)
    ax1.plot(t0,mej,'k.')
    ax2.plot(t0,ek,'k.')
ax1.grid(True)
ax1.set_xlabel('T$_\mathrm{0 \;}$ (d)',fontsize=12)
ax3.set_xlabel('R$_{decline}$ (mag/100d)',fontsize=12)
ax2.set_ylabel('E$_\mathrm{kin \;}$ (foe)',fontsize=12)
ax1.set_ylabel('M$_{ej}$ (M$_\odot$)',fontsize=12)

# set y axis
mej1, mej2 = 1, 12
ax1.set_ylim([mej1, mej2])
def mej_to_ek(mej, vej): return (vej*1e8)**2*(mej*constants.M_sun)/2/1e51
ax2.set_ylim([mej_to_ek(mej1, vej), mej_to_ek(mej2, vej)])
for _ in [ax1,ax2]: _.set_xlim(40,110)
ax3.set_xlim(3.20, 1.96)
```

### Tail model, t0 (slope) vs bolometric mag at 60 d post peak

```python
fig,ax=plt.subplots(1,1)
trise=-13.14
x,y=[],[]
xx = 60
for mni in np.arange(1e-3,2,1e-3):
    yy = functions.Lbol_to_Mbol(models.arnett_tail.tail_fit_t0(xx, mni, 50))
    x.append(mni)
    y.append(yy)
ax.plot(x,y,'r-')
ax.set_xlabel('M$_\mathrm{Ni}$ (M$_\odot$)',fontsize=12)
ax.set_ylabel('M$_\mathrm{bol}$ at 60d',fontsize=12)
ax.invert_yaxis()
ax.set_xscale('log')
plt.grid(True)
```

### Tail model, t0 vs slope

```python
fig,ax=plt.subplots(1,1)
trise=-13.14
x,y=[],[]
xx = np.arange(40,120,1)
for t0 in np.arange(0,300,1):
    yy = functions.Lbol_to_Mbol(models.arnett_tail.tail_fit_t0(xx, 0.09, t0))
    ss = (max(yy)-min(yy)) / (max(xx)-min(xx)) * 100.
    x.append(t0)
    y.append(ss)
    if t0 in [40,100]: print (t0, ss)
ax.plot(x,y,'r-')
ax.plot([0,300],[.98,.98],'k--') # colbat decay, s=0.98 mag/100d
ax.set_xlabel('T$_\mathrm{0 \;}$ (d)',fontsize=12)
ax.set_ylabel('R$_{decline}$ (mag/100d)',fontsize=12)
ax.set_xlim([0,300])
plt.grid(True)
```

### Tail model, Mej, Ek vs t0

```python
xs = np.arange(.1,50,.1) # Mej
ys = np.arange(.1,50,1)  # Ekin
x, y = np.meshgrid(xs, ys)
z = models.arnett_tail.Mej_Ek_to_t0(x, y)
levels = np.arange(0,1000,100)

fig, ax = plt.subplots(constrained_layout=True)
CS = ax.contourf(x, y, z, levels, extend='both')
CS2 = ax.contour(CS, levels, colors=['r','g','y','b','k','orange','grey','cyan','brown'], origin='lower')

ax.set_xlabel('M$_{ej}$ (M$_{\odot}$)')
ax.set_ylabel('E$_{k}$ (foe)')

for v, c in zip([5, 10, 15, 20], ['r', 'b', 'g', 'k']):
    Ek = (v * 1e8)**2 * xs / 2. * constants.M_sun * 1e-51
    ax.plot(xs, Ek, ls='--', color=c, label='v$_{m}$=%s e3 km/s'%v)

cbar = fig.colorbar(CS)
cbar.ax.set_ylabel(r'$t_0$ (days)')
cbar.add_lines(CS2)
ax.set_yscale('log')
ax.set_xlim([.1,50])
ax.set_ylim([.1,50])
ax.legend()
```

### Arnett model, Mej, Ek vs taum

```python
xs = np.arange(.1,10,.1)
ys = np.arange(.1,100,1)
x, y = np.meshgrid(xs, ys)
z = models.arnett_tail.Mej_Ek_to_taum(x, y)

levels = np.arange(0,100,10)

fig, ax = plt.subplots(constrained_layout=True)
CS = ax.contourf(x, y, z, levels, extend='both')
CS2 = ax.contour(CS, levels, colors=['r','g','y','b','k','orange','grey','cyan','brown'], origin='lower')

ax.set_xlabel('M$_{ej}$ (M$_{\odot}$)')
ax.set_ylabel('E$_{k}$ (foe)')

for v, c in zip([5, 10, 15, 20], ['r', 'b', 'g', 'k']):
    Ek = (v * 1e8)**2 * xs / 2. * constants.M_sun * 1e-51
    ax.plot(xs, Ek, ls='--', color=c, label='v$_{m}$=%s e3 km/s'%v)

cbar = fig.colorbar(CS)
cbar.ax.set_ylabel(r'$\tau_m$ (days)')
cbar.add_lines(CS2)
ax.set_yscale('log')
ax.set_xlim([.1,10])
ax.set_ylim([.1,100])
ax.legend()
```
