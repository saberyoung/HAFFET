# run HAFFET with a list of Ibc SNe

### import library

```python
from sdapy import snerun
import matplotlib.pyplot as plt
plt.rcParams["text.usetex"] = False # disable latex
import numpy as np
```

### initialize a snobject class with proper meta data

```python
sncls = snerun.snobject(objid='ZTF20aajcdad', aliasid='SN2020bcq',
    z=0.0186, dm=34.6, mkwebv=0.01387, hostebv=0, sntype='Ib', 
    ra='13:26:29.65', dec='+36:00:31.1', jdpeak=2458888.02)
```

### coordinates to degrees

```python
sncls.parse_coo(deg=True)
```

### coordinates to healpix indices

```python
sncls.parse_coo(deg=False, hpx=True, nside=512)
```

### query current julian date

```python
sncls.mjd_now(jd=True)
```

```python
sncls.mjd_now(jd=False)
```

### query ZTF alert LC via ztfquery

```python
sncls.query_alert_ztf(source='marshal')
```

### get local ZTF alert LCs

```python
sncls.get_alert_ztf(source='marshal')
print (sncls.lc.query("source=='marshal'").iloc[0])
```

### query ATLAS forced photometry

```python
sncls.query_fp_atlas()
```

### get local ATLAS forced photometry

```python
sncls.get_fp_atlas(binDays=1, clobber=False)
print (sncls.lc.query('source == "atlasfp"').iloc[0])
```

### query ZTF forced photometry

```python
sncls.query_fp_ztf(get_email=False)
```

### get local ZTF forced photometry

```python
sncls.get_fp_ztf()
print (sncls.lc.query('source == "ztffp"').iloc[0])
```

### query Open Astronomy Catalog LCs

```python
sncls.query_oac(db='AL', which='photometry')
```

### get local Open Astronomy Catalog LCs
### note: this object is not included, thus get empty table

```python
sncls.get_oac(which='photometry')
print (sncls.lc.query('source == "oac"'))
```

### Let's try another famous one from Open Astronomy Catalog

```python
_sncls = snerun.snobject(objid='1998bw')
_sncls.query_oac(db='AL', which='photometry')
_sncls.get_oac(which='photometry')
print (_sncls.lc.query('source == "oac"').iloc[0])
```

### query spectra from fritz

```python
sncls.query_spectra(source='fritz')
```

### load spectra

```python
sncls.get_local_spectra()
print (sncls.spec.data)
```

### astro rapid to guess SN type based on multi-band LCs

```python
fig, ax = plt.subplots(1,1)
sncls.rapid(ax=ax)
```

### correct ZTF CCD baselines

```python
fig,ax=plt.subplots(1,1)
baseline = sncls.calibrate_baseline(ax=ax, key='fcqfid', source='ztffp', xmin=-100, xmax=-20)
sncls.correct_baseline(baseline, key='fcqfid', source='ztffp')
```

### OK, let's go to fitting part
### first, do Gaussian Process interpolation

```python
sncls.run_gp(gp_bands=['r', 'g', 'i', 'c', 'o'])
sncls.set_peak_gp('r') # set t0 with GP in r band
```

### Plot them

```python
sncls.fig,(sncls.ax, sncls.ax2)=plt.subplots(2,1)
sncls._ax(plot_bands=['r', 'g', 'i', 'c', 'o'], show_title=False)
sncls._ax2(plot_bands=['r', 'g', 'i', 'c', 'o'])
```

```python
sncls.clip_lc(clipsigma=5)
sncls.fig, sncls.ax=plt.subplots(1,1)
#sncls.set_texp_pl('r', model_name=None)
sncls._ax(plot_bands=['r','g','i','c','o'], show_title=False, ax_ylim=(0,200), ax_xlim=(-30,0))
```

### Then we can run different fittings

### first what engines are available?

```python
from sdapy.model_fitters import get_engine
get_engine()
```

### OK, let's do multiband_main fittings first
### what models are available then?

```python
from sdapy.model_fitters import get_model
get_model(which_engine='multiband_main', with_alias=True)
```

### OK, let's see what bazin model fits?

```python
from sdapy.model_fitters import get_pars
get_pars('bazin', with_alias=True)
```

### OK, let's fitting with bazin model, via emcee
### note, for the second time, set fit_redo to False, you will easily get all the samplings

```python
sncls.run_fit('multiband_main',
            fit_methods = ['bazin'],
            multiband_main_bands = ['i','g','r'],
            multiband_main_xrange = [-50, 50], 
            multiband_main_xrangep = [-50, 50], 
            multiband_main_routine = 'mcmc',
            nwalkers = 30,
            nsteps = 20000,
            nsteps_burnin = 2000,
            fit_redo = False,
            source = None,
           )
```

### let's show the fitting contours
### set clobber to Ture if corner of the models is previously done

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))
sncls.show_corner(ax, index=0, gp=False, engine=None, model='bazin', source=None, filts=['r'], clobber=False)
```

### Plot them

```python
sncls.summarize_plot(ax=None)
```

### run the same fits with scipy.optimize.minimize will take less time

```python
sncls.run_fit('multiband_main',
            fit_methods = ['bazin'],
            multiband_main_bands = ['i','g','r'],
            multiband_main_xrange = [-50, 50], 
            multiband_main_xrangep = [-50, 50], 
            multiband_main_routine = 'minimize',
            fit_redo = True,
           )
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))
sncls.show_corner(ax, index=0, gp=False, engine=None, model='bazin', source=None, filts=['r'], clobber=False)
```

### OK, then let's go to multiband_early fittings

```python
get_model(which_engine='multiband_early', with_alias=True)
```

### this is actually similar as what was done in Millar et al 2018, i.e. fit all bands with power law together, for a common texp

```python
get_pars('powerlaw_multiple', with_alias=True)
```

### OK, let's fit them

```python
sncls.run_fit(
        'multiband_early',
        fit_methods = ['powerlaw_multiple'],
        multiband_early_bands = ['g','r','i','c','o'],
        multiband_early_xrange = [-50,0],
        multiband_early_xrangep = [-50,0],
        multiband_early_routine = 'mcmc',
        nwalkers = 200,
        nsteps = 20000,
        nsteps_burnin = 10000,
        fit_redo = True,
        source = None,
)
```

### OK, let's see the contours

```python
fig, ax = plt.subplots(1,1,figsize=(15,15))
sncls.show_corner(ax, index=0, gp=False, engine=None, model='powerlaw_multiple', source=None,clobber=False)
```

### reproduce the early LCs
