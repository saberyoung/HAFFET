```python
df=pd.read_csv('/Users/yash0613/Desktop/scripts/sn_data_analysis_latest/data/SN2018bsz_SDAUBgcVroizJHKWX.txt',
                sep='\s*',).fillna(99).drop_duplicates()
rows = []
for _ in df.index:
    for f in df.keys():
        if f == 'mjd' or '_err' in f: continue
        mjd = df['mjd'][_]
        if mjd[0] != '#' and df[f][_] != 99:
            m, me = df[f][_], df['%s_err'%f][_]
            if f == 'UVW1': f='A'
            elif f == 'UVM2': f='D'
            elif f == 'UVW2': f='S'
            elif f == 'W1': f='W'
            elif f == 'W2': f='Q'
            if f not in filters.central_wavelengths: print (f)
            rows.append((mjd, f, m, me))
df = pd.DataFrame(rows, columns=['mjd','filter','mag','emag'])
for k in ['mjd','mag','emag']: df[k] = pd.to_numeric(df[k])
df['jdobs'] = df['mjd'] + 2400000.5
```

```python
from scipy.optimize import curve_fit
from scipy import integrate

SN_distance = 10**((ztfp.dm - 25)/5) # Mpc
SN_distance *= 3.086e24 # convert Mpc to cm, since flux in erg/s/cm2/A
z = ztfp.z
t0 = ztfp.t0
do_Kcorr=True

fig, axarr = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))
plt.subplots_adjust(wspace = 0., hspace = 0.)

j = 0
jdl, Ll, eLl, Tl, eTl, Rl, eRl, L1l, eL1l = [], [], [], [], [], [], [], [], []
for jd in np.unique(df['jdobs']):
    _df = df.query('jdobs==@jd')
    if len(_df) == 1: continue
    phase = (jd - t0)/(1+z)
    if phase < 150: continue
    
    ax = axarr[int(j/2), j%2]
    # flux in various filters
    fluxes, efluxes, ws, bs = [], [], [], []
    for filt in ['H', 'J', 'K', 'W', 'Q']:
        __df = _df.query('filter==@filt')
        if len(__df) == 0: 
            xx,yy,yye,ff = ztfp.gpcls.predict(x_pred=[jd], returnv=True,)
            __ = np.where(ff==filt)
            xx = xx[__]
            yy = yy[__]
            yye = yye[__]
            print (yy)
            if len(xx) == 0: continue
            _mag = ztfp.flux_to_mag(yy, dflux=None, sigma=1, zp=23.9)[0]
            _emag = abs(ztfp.flux_to_mag(yy, dflux=None, sigma=1, zp=23.9)[0]
                     -ztfp.flux_to_mag(yy+yye, dflux=None, sigma=1, zp=23.9))[0]
            if np.isnan(_mag):continue
        else:
            _mag = float(__df['mag'])
            _emag = float(__df['emag'])
        
        # If UVOT bands are in AB, need to convert to Vega
        if filt in filters.ab_to_vega.keys():
            _mag -= filters.ab_to_vega[filt]
        
        wlref = filters.central_wavelengths[filt]
        fref = filters.zp[filt] * 1e-11
        bandwidths = filters.filter_width[filt]
        if do_Kcorr:
            wlref /= (1+z)
            fref *= (1+z)
            bandwidths /= (1+z)
        _flux = 4*np.pi*SN_distance**2*fref*10**(-0.4*_mag)
        _dflux = 2.5/np.log(10) * _flux * _emag
        fluxes.append(_flux)
        efluxes.append(_dflux)
        ws.append(wlref)
        bs.append(bandwidths)                      
        ax.errorbar(wlref, _flux, yerr = _dflux, xerr=bandwidths, **filters.PROP1[filt])
    fluxes, efluxes, ws, bs = np.array(fluxes), np.array(efluxes), np.array(ws), np.array(bs)
    ax.set_xlim([-100, 5e4])
    ax.set_ylim([1e30, 3e37])
    ax.set_yscale('log')
    
    # bb fit
    BBparams, covar = curve_fit( functions.bbody, ws, fluxes/1e40, p0=(10,2),
                        sigma=efluxes/1e40, absolute_sigma=True, maxfev=20000)
    perr = np.sqrt(np.diag(covar))
    popt_up = BBparams + perr
    popt_dw = BBparams - perr
    
    _T = BBparams[0]
    _Te = min(np.sqrt(np.diag(covar))[0],_T)
    _R = np.abs(BBparams[1])
    _Re = min(np.sqrt(np.diag(covar))[1],_R)
    Tl.append(_T)
    eTl.append(_Te)
    Rl.append(_R)
    eRl.append(_Re)                                                               
    jdl.append(phase)

    # Get pseudobolometric luminosity by trapezoidal integration,
    #    with flux set to zero outside of observed bands
    _L = integrate.trapz(fluxes[np.argsort(ws)],ws[np.argsort(ws)])
    Ll.append(_L)
                
    # Use flux errors and bandwidths to get luminosity error
    _Le = np.sqrt(np.sum((bs*efluxes)**2))
    eLl.append(_Le)
                
    # Calculate luminosity using alternative method of Stefan-Boltzmann, and T and R from fit
    _Lbb = 4*np.pi*(_R*1e15)**2*5.67e-5*(_T*1e3)**4
    _Lbbe = _Lbb*np.sqrt((2*_Re/_R)**2+(4*_Te/_T)**2)
    L1l.append(_Lbb)
    eL1l.append(_Lbbe)                
    
    if phase < 150: w1 = np.arange( 0,3e4,1. )
    else: w1 = np.arange( 0,5e4,1. )
    fit = functions.bbody(w1, *BBparams) * 1e40
    fit_up = functions.bbody(w1, *popt_up) * 1e40
    fit_dw = functions.bbody(w1, *popt_dw) * 1e40
    
    ax.plot(w1, fit, 'k--', lw=1)
    ax.fill_between(
            w1, fit_up, fit_dw, alpha=0.2, color='k')
    
    #if j==9: ax.legend(loc = 'center left', bbox_to_anchor = (1.0, 0.5), ncol=1, numpoints=1, fontsize=12, frameon=True)

    #if j % 2 == 0 : ax.set_yticks([1e33, 1e35, 1e37, 1e39])
    #else:     ax.set_yticks([])
    if j == 2: ax.set_xticks([1e3, 4e4])
    else:     ax.set_xticks([])
    
    #ax.text(1.2e4, 1e40, '%.1f d' % (phase) )
    j += 1
    
#axarr[3, 0].set_ylabel('Luminosity (erg s-1 A-1)')
#axarr[6, 4].set_xlabel(r'             Wavelength ($\AA$)')
#axarr[10, 9].set_visible(False)

#plt.show()
#p, L, dL1, dL2, T, dT, R, dR1, dR2 = np.array(p), np.array(L), np.array(dL1), np.array(dL2), np.array(T), np.array(dT), np.array(R), np.array(dR1), np.array(dR2)
#L1s, L2s = np.array(L1s), np.array(L2s)
#fig.savefig('bbfit_gri.png', bbox_inches='tight', dpi=400)
```
